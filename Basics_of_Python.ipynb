### Sum of digit 

```python
num = int(input("Enter 3 digit number "))
first_digit = num // 100 #return only decimal value
second_digit = (num//10)%10
last_digit = num % 10 #gives reminder after divison
sum = first_digit + second_digit + last_digit
print("This is the sum of first and last digit",sum)
```

### Sum of 1 and last digit

```python
num = int(input("Enter 3 digit number "))
first_digit = num // 100 #return only decimal value
last_digit = num % 10 #gives reminder after divison
sum = first_digit + last_digit
print("Your first number is",first_digit)
print("Your last number is",last_digit)
print("This is the sum of first and last digit",sum)
```
