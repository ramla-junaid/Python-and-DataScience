```python
name = str(input("Enter student name : "))
fname = str(input("Enter student father name : "))
standard = int(input("Enter student class : "))
print("Enter marks of following subjects")
english = int(input("Enter english marks : "))
urdu = int(input("Enter urdu marks : "))
maths = int(input("Enter maths marks : "))
science = int(input("Enter science marks : "))
computer = int(input("Enter computer marks : "))

first_letter_of_student_name = name[0]
last_letter_of_student_father_name = fname[-1]
roll_numb = (first_letter_of_student_name + last_letter_of_student_father_name) +str(standard)
percentage=((english+urdu+maths+science+computer)/500)*100

print("---------Report Card---------")
print("Name : ",name)
print("Father name : ",fname)
print("Class : ",standard)
print("Roll Number :",roll_numb)
print("English : ",english)
print("Urdu : ",urdu)
print("Maths : ",maths)
print("Science : ",science)
print("Computer : ",computer)
print("Percentage : ",percentage)
```
