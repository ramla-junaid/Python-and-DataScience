### take 3 input from user and store them in list and the add one more using append

```python
num1 = int(input("Enter 1 num "))
num2 = int(input("Enter 2 num "))
num3 = int(input("Enter 3 num "))
numlist = [num1, num2, num3]
print("List before append",numlist)
num4 = int(input("Enter 4 num "))
numlist.append(num4)
print("List after append",numlist)
```

### Report Card

```python
name = input("Enter your name")
fname = input("Enter father name")
clas = input("Enter your class")
eng = int(input("Enter english marks"))
urdu = int(input("Enter urdu marks"))
math = int(input("Enter maths marks"))
sci = int(input("Enter science marks"))
total = eng+urdu+math+sci
percentage=(total/400)*100
dict = {"name":name ,"fname":fname ,"class":clas,"total":total,"per":percentage}
print("REPORT CARD")
print("Name :",dict["name"])
print("Father Name :",dict["fname"])
print("Class :",dict["class"])
print("Your total marks are :",dict["total"])
print("Your percentage is :",dict["per"])
```
